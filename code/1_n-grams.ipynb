# 1. Bibliotecas

```
import numpy as np 
import math 
import pandas as pd 

#Para iteradores y diccionarios /frames
import collections 
from itertools import permutations, zip_longest, islice, tee
from tabulate import tabulate
```

# 2. Import datasets

Importamos los datos y los hacemos compatibles con las respectivas bibliotecas 

English dataset


```
#en = open("/content/derechos-ingles.txt")
#countEn = collections.Counter(en)
#print(countEn)

en = pd.read_csv('/content/derechos-ingles.txt', header = None, delimiter = "\t", names=["text"])
en = en.dropna()
en['text'] = en['text'].astype(str) 
en = ' '.join(en["text"])
```

Español dataset

```
es = pd.read_csv('/content/derechos-espanol.txt', header = None, delimiter = "\t", names=["text"])
es = es.dropna()
es['text'] = es['text'].astype(str) 
es = ' '.join(es["text"])
```

Francés dataset


```
fr = pd.read_csv('/content/derechos-frances.txt', header = None, delimiter = "\t", names=["text"])
fr = fr.dropna()
fr['text'] = fr['text'].astype(str) 
fr = ' '.join(fr["text"])
```

# 3. Get Bigrams



```
n= 2 # escogeremos bigramas como partida
```

```
en_grams = zip_longest(*(islice(seq, index, None) for index, seq in enumerate(tee(en, n))))
en_grams = list(en_grams)
print(en_grams)
```

```
es_grams = zip_longest(*(islice(seq, index, None) for index, seq in enumerate(tee(es, n))))
es_grams = list(es_grams)
print(es_grams)
```

```
fr_grams = zip_longest(*(islice(seq, index, None) for index, seq in enumerate(tee(fr, n))))
fr_grams = list(fr_grams)
print(fr_grams)
```

# 4. N-grams  counter



## 4.1 INGLÉS

```
gramCountEN = collections.Counter(en_grams)
#print("All grams:\n " + str(gramCount))
commonGramEN = gramCountEN.most_common(30)
leastGramEN = gramCountEN.most_common()[:-30-1:-1]
print("Más común\n")
print(tabulate(commonGramEN, headers=['n-gram', 'cantidad'], tablefmt='orgtbl'))
```

```
print("Menos común\n")
print(tabulate(leastGramEN, headers=['n-gram', 'cantidad'], tablefmt='orgtbl'))
```

## 4.2 Español

```
gramCountES = collections.Counter(es_grams)
#print("All grams:\n " + str(gramCount))
commonGramES = gramCountES.most_common(30)
leastGramES = gramCountES.most_common()[:-30-1:-1]
print("Más común\n")
print(tabulate(commonGramES, headers=['n-gram', 'cantidad'], tablefmt='orgtbl'))
```

```
print("Menos común\n")
print(tabulate(leastGramES, headers=['n-gram', 'cantidad'], tablefmt='orgtbl'))
```

## 4.3 FRENCH


```
gramCountFR = collections.Counter(fr_grams)
#print("All grams:\n " + str(gramCount))
commonGramFR = gramCountFR.most_common(30)
leastGramFR = gramCountFR.most_common()[:-30-1:-1]
print("Más común\n")
print(tabulate(commonGramFR, headers=['n-gram', 'cantidad'], tablefmt='orgtbl'))
```

```
print("Menos común\n")
print(tabulate(leastGramFR, headers=['n-gram', 'cantidad'], tablefmt='orgtbl'))
```
